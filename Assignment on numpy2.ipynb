```python
import numpy as np
import matplotlib.pyplot as plt

#A Histogram is a graph showing frequency distributions.
#It is a graph showing the number of observations within each given interval
#they are graphical representation that organises a group of dats points in a specified range


x=[1,12,22,21,20,21,31,2,32,40,14,33,50,44,
  45,46,44,43,45,43,42,41,40,43,45,43,44,61,
  70,12,50,60,55,43,32,89,90,54,45,57,
  32,12,18,85,25,34,38,41,52,52,52,18,12,
  32,89,85,54,20,23,25,27,32,43]
plt.hist(x,5,range=(0,100),ec="r")
 
plt.xlabel("age")
plt.ylabel("count")
plt.title("population age count")
plt.grid(0)

plt.show()
```

# SCATTER PLOT


```python
#A scatter plot is a diagram where each value in the data set is represented by a dot.
#The matplotlib module has a method for drawing scatter plot,
#it needs two arrays of same length,
#one for the value of x-axis,
#one for the value of y-axis.


from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[10,50,100,80,60,90]
color=["red","yellow","blue","pink","orange","purple"]
size=[10,100,20,90,50,70]

plt.scatter(x,y,c=color,s=size)

plt.show()
```

# BOXPLOT

```python
# A Box Plot is also known as Whisker plot is created to display the summary of the set of data values
# having properties like minimum,firt quartile,
# median,third quartile and maximum.
# A box plot is a graphical and standard way to display,
# the distribution of data using 5keys numbers


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
np.random.seed(10)
data = np.random.normal(100, 20, 200)
 
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
plt.boxplot(data)
 
# show plot
plt.show()
```

# AGGREGATE FUNCTIONS

```python
# Aggregate functions perform an operation on set of values and produce a single result.
# Aggregate function is array wise sum
# Aggregate function is getting a single numerical output from n number of numerical elements.
# It is a function where the values of multiple rows are grouped together to form a single summary value.


import numpy as np
a= np.array([1,2,3])
print(a)
```

```python
b= np.array([(1.5,2,3), (4,5,7)])
print(b)
```

```python
array = np.random.randint(16,size=(4,4))
print(array)

#argmin
print("indices of min element in y axis: ", np.argmin(array,axis=0))
print("indices of min element in x axis: ", np.argmin(array,axis=1))


#argmax
print("indices of min element in y axis: ", np.argmax(array,axis=0))
print("indices of min element in x axis: ", np.argmax(array,axis=1))
```

## UNIVERSAL FUNCTION

```python
#UNFUNC(UNIVERSAL FUNCTION)
#it is quite faster than a normal loop
# is aunction that operates on ndarrays in an element by element function 
# supporting array broadcasting, type casting and several other standard features
# they are a powerful tool for performing mathematical operations on Numpy arrays.
```

```python
import numpy as np
list1 =[5,15,25,30]
list2 =[2,4,7,9]
z =[]
for a,j in zip(list1, list2):
    print(a+j)
```
